```python
class person:
  def __init__(self, nama, umur):
    self.nama = nama
    self.umur = umur

  def tampilkan_info(self):
    print(f"Nama: {self.nama}, Umur: {self.umur}")
```

```python
class Mahasiswa(person):
  def __init__(self, nama, umur, nim):
    super().__init__(nama, umur)
    self.nim = nim

  def tampilkan_mahasiswa(self):
    self.tampilkan_info()
    print(f"NIM: {self.nim}")

class Dosen(person):
  def __init__(self, nama, umur, nidn):
    super().__init__(nama, umur)
    self.nidn = nidn

  def tampilkan_dosen(self):
    self.tampilkan_info()
    print(f"NIDN: {self.nidn}")

mhs = Mahasiswa("Fadil", 20, "D12345")
mhs.tampilkan_mahasiswa()

dsn = Dosen("Dr. Budi", 40, "N67890")
dsn.tampilkan_dosen()
```

```python
class Barang:
  def __init__(self, id_barang, nama, harga):
    self.id_barang = id_barang
    self.nama = nama
    self.harga = harga

  def info(self):
    return f"[{self.id_barang}] {self.nama:15} | Rp{self.harga:10,}"
```

```python
class BarangElektronik(Barang):
  def __init__(self, id_barang, nama, harga, garansi):
    super().__init__(id_barang, nama, harga)
    self.garansi = garansi

  def info(self):
    return f"{super().info()} | Garansi: {self.garansi} bln"

class BarangKonsumsi(Barang):
  def __init__(self, id_barang, nama, harga, tgl_exp):
    super().__init__(id_barang, nama, harga)
    self.tgl_exp = tgl_exp

  def info(self):
    return f"{super().info()} | Exp: {self.tgl_exp}"

laptop = BarangElektronik("E01", "Laptop Gaming", 15000000, 24)
susu = BarangKonsumsi("K05", "Susu UHT", 18000, "2026-12-01")

print(laptop.info())
print(susu.info())
```

```python
class Kendaraan:
  def __init__(self, id_kendaraan, merk, kecepatan):
    self.id_kendaraan = id_kendaraan
    self.merk = merk
    self.kecepatan = kecepatan

  def info(self):
    print(f"ID: {self.id_kendaraan}, Merk: {self.merk}, Kecepatan: {self.kecepatan}")
```

```python
class Mobil(Kendaraan):
  def __init__(self, id_kendaraan, merk, kecepatan, jumlah_pintu):
    super().__init__(id_kendaraan, merk, kecepatan)
    self.jumlah_pintu = jumlah_pintu

  def info(self):
    super().info()
    print(f"Jumlah Pintu: {self.jumlah_pintu}")

class Motor(Kendaraan):
  def __init__(self, id_kendaraan, merk, kecepatan, jenis):
    super().__init__(id_kendaraan, merk, kecepatan)
    self.jenis = jenis

  def info(self):
    super().info()
    print(f"Jenis: {self.jenis}")

mobil1 = Mobil("M001", "Toyota", 120, 4)
motor1 = Motor("M002", "Honda", 100, "Sport")

print(mobil1.info())
print(motor1.info())
```

```python
#Latihan 1
class Hewan:
    def __init__(self, nama, habitat):
        self.nama = nama
        self.habitat = habitat

    def info(self):
        return f"{self.nama} hidup di {self.habitat}"

class Kucing(Hewan):
    def __init__(self, nama, habitat, warna_bulu, jenis_makanan):
        super().__init__(nama, habitat)
        self.warna_bulu = warna_bulu
        self.jenis_makanan = jenis_makanan

    def info(self):
        return super().info() + f", bulunya {self.warna_bulu}, makan {self.jenis_makanan}"

class Paus(Hewan):
    def __init__(self, nama, habitat, ukuran, jenis_paus):
        super().__init__(nama, habitat)
        self.ukuran = ukuran
        self.jenis_paus = jenis_paus

    def info(self):
        return super().info() + f", ukurannya {self.ukuran}, jenis {self.jenis_paus}"

class Singa(Hewan):
    def __init__(self, nama, habitat, jumlah_kaki, kekuatan):
        super().__init__(nama, habitat)
        self.jumlah_kaki = jumlah_kaki
        self.kekuatan = kekuatan

    def info(self):
        return super().info() + f", berkaki {self.jumlah_kaki}, kekuatannya {self.kekuatan}"

kucing = Kucing("Kucing Persia", "Rumah", "Putih", "Ikan")
paus = Paus("Paus Biru", "Laut", "Sangat Besar", "Biru")
singa = Singa("Singa Afrika", "Padang Rumput", 4, "Sangat Kuat")

print(kucing.info())
print(paus.info())
print(singa.info())
```

```python
#Latihan 2
class Produk:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga

    def hitung_diskon(self):
        return self.harga  # tanpa diskon

    def tampilkan(self):
        print("Nama Produk :", self.nama)
        print("Harga Asli  :", self.harga)
        print("Harga Diskon:", self.hitung_diskon())
        print("-" * 30)

class Elektronik(Produk):
    def __init__(self, nama, harga, garansi):
        super().__init__(nama, harga)
        self.garansi = garansi

    def hitung_diskon(self):
        return self.harga * 0.10  # diskon 10%

class Pakaian(Produk):
    def __init__(self, nama, harga, ukuran):
        super().__init__(nama, harga)
        self.ukuran = ukuran

    def hitung_diskon(self):
        return self.harga * 0.20  # diskon 20%


produk1 = Elektronik("Laptop", 10000000, "2 Tahun")
produk2 = Pakaian("Kaos", 100000, "L")

produk1.tampilkan()
produk2.tampilkan()
```
